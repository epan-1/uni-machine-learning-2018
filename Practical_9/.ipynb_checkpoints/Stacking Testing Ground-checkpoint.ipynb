```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_predict
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
y = iris.target

def convert_class(raw):
    if int(raw)<=10: 
        return 0
    else: 
        return 1

# Open up the Abalone dataset

f = open('../Practical_8/abalone.csv', 'r')
X = []
y = []

for line in f:
    atts = line[:-1].split(",")
    X.append(atts[1:-1])
    y.append(convert_class(atts[-1]))

f.close()

# Convert to numpy array
X = np.array(X).astype(np.float)

# X = ['a', 'b', 'b', 'a', 'a', 'b']
# y = [1, 0, 0, 1, 0, 0]
n_folds = 10

final_clf = LogisticRegression()
clfs  = [DecisionTreeClassifier(), LinearSVC(), MultinomialNB()]

kf = KFold(n_splits=n_folds)
output = []

temp_data = pd.DataFrame()
# Column index
col = 0

predictions = [[] for i in range(len(clfs))]
y_labels = []
    
    
for train, test in kf.split(X, y):
    new_X = [X[i] for i in train]
    new_y = [y[i] for i in train]
    new_test = [X[i] for i in test]


    for train_2, test_2 in kf.split(new_X, new_y):
        X_train = [new_X[i] for i in train_2]
        y_train = [new_y[i] for i in train_2]
        clf_test = [new_X[i] for i in test_2]
        clf_y = [new_y[i] for i in test_2]
        
        classifier_num = 0
        for classifier in clfs:
            classifier.fit(X_train, y_train)
            # Also predict the training instance and add it to the column
            # for the new dataset
            predictions[classifier_num] += classifier.predict(clf_test).tolist()
            classifier_num += 1
        y_labels += clf_y

# Create the new dataset
temp_data[0] = predictions[0]
temp_data[1] = predictions[1]
temp_data[2] = predictions[2]
temp_data[3] = y_labels

final_train = temp_data.iloc[:,0:-1].as_matrix()
final_y = temp_data.iloc[:, -1].as_matrix()

# print(np.average(cross_val_score(final_clf, final_train, final_y, cv=n_folds)))
final_clf.fit(final_train, final_y)
final_clf.predict(X)

lr = LogisticRegression()
print(np.average(cross_val_score(lr, X, y, cv=n_folds)))

nb = LinearSVC()
print(np.average(cross_val_score(nb, X, y, cv=10)))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_predict
from sklearn import datasets

# iris = datasets.load_iris()

# X = iris.data
# y = iris.target

n_folds = 10

final_clf = LinearSVC()
clfs  = [MultinomialNB(), LogisticRegression(), DecisionTreeClassifier()]

kf = KFold(n_splits=n_folds)
output = []

temp_data = pd.DataFrame()
# Column index
col = 0

for classifier in clfs:
    
    predictions = []

    for train, test in kf.split(X, y):
        new_X = [X[i] for i in train]
        new_y = [y[i] for i in train]
        new_test = [X[i] for i in test]
        
        classifier.fit(new_X, new_y)
        predictions += classifier.predict(new_test).tolist()
        
        
    temp_data[col] = predictions
    col += 1

temp_data[col] = y

final_train = temp_data.iloc[:,0:-1].as_matrix()
final_y = temp_data.iloc[:, -1].as_matrix()

print(np.average(cross_val_score(final_clf, final_train, final_y, cv=n_folds)))
nb = LinearSVC()
print(np.average(cross_val_score(nb, X, y, cv=10)))
```
